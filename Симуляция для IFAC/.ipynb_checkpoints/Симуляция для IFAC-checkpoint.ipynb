```python
import pygame
import numpy as np
import numpy.linalg as la
import random as rdm
import math as mh
import math
import time
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd

FPS = 50
SForFrame = 1 / FPS
WIN_WIDTH = 1000
WIN_HEIGHT = 600
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
BRAWN = (101, 67, 33)
GRAY = (125, 125, 125)
LIGHT_BLUE = (64, 128, 255)
GREEN = (0, 128, 0)
RED = (255, 0, 0)
LIME = (0, 255, 0)
YELLOW = (225, 225, 0)
ORANGE = (255, 165, 0)
PINK = (230, 50, 230)
BLUE = (0, 0, 150)
BGCOLOR = WHITE#BLUE
MINROBOTCOLOR = (80, 80, 80)
MAXROBOTCOLOR = (200, 200, 200)
ROBOTCOLOR = (200, 200, 200)
WALLCOLOR = BLACK
LINECHARTCOLOR = GRAY

class VisualObj:
    def draw(self):
        pass

class ActiveObj(VisualObj):
    def action(self):
        pass
    
class GeomObj(ActiveObj):
    def getDist(self, oPos):
        pass
    def getXDiff(self, oPos):
        pass
    def getYDiff(self, oPos):
        pass
    def getYDist(self, oPos):
        return abs(self.getYDiff(oPos))
    def isAbove(self, oPos):
        return self.getYDiff(oPos) > 0
    
class Robot(GeomObj):
    def __init__(self, pos, V = np.array([0, 0]), Rvis = 10, r = 10, color = ROBOTCOLOR, Dir = np.array([1, 0]), maxAngVel = 14, maxAcc = 0.08): #maxAngVel = 14, maxAcc = 0.08
        self._dir = Dir
        self._live = True
        self._pos = pos
        self._maxAngVel = maxAngVel / 180 * mh.pi
        self._maxAcc = maxAcc
        self._V = V
#         if la.norm(V) > 1e-2: #Костыль
#             self._V = V
#         else:
#             self._V = Dir
        self._r = r
        self._color = color
        self._state = 'ordinary'
    def setState(self, newState):
        self._state = newState
    def getState(self):
        return self._state
    def getColor(self):
        return self._color
    def rise(self):
        self._live = True
    def kill(self):
        self._live = False
    def isLive(self):
        return self._live
    def teleportation(self, tPos):
        self._pos = tPos
    def getPos(self, distortion = 5.1): #5.1
        if distortion < 1:
            return self._pos
        else:
            return self.getDistoredPos(distortion)
    def getDistoredPos(self, R = 7):
        return self._pos + getRndCrcPnt(R)
    def getDir(self):
        return self._dir
    def getOrtDir(self):
        return np.array([self._dir[1], -self._dir[0]])
    def setDir(self, nDir):
        self._dir = nDir / la.norm(nDir)
    def getDist(self, oPos):
        return la.norm(self._pos - oPos)
    def getXDiff(self, oPos):
        return self._pos[0] - oPos[0]
    def getYDiff(self, oPos):
        return self._pos[1] - oPos[1]
    def getDirDiff(self, oPos, Dir):
        u = self._pos - oPos
        return u[0] * Dir[0] + u[1] * Dir[1]
    def getRDirDiff(self, oPos):
        return self.getDirDiff(oPos, self.getDir())
    def getOrtRDirDiff(self, oPos):
        return self.getDirDiff(oPos, self.getOrtDir())
    def isDirAbove(self, oPos, oDir):
        return self.getDirDiff(oPos, oDir) < 0
    def inRank(self, oPos, buildWidth):
        return abs(self.getXDiff(oPos)) < buildWidth
    def setV(self, V):
        scal = V[0]*self._V[0]+V[1]*self._V[1];
        vect = V[0]*self._V[1]-V[1]*self._V[0];
        ang = mh.atan2(vect, scal)
        absV = la.norm(V)
        oldV = la.norm(self._V)
        if abs(ang) < self._maxAngVel:
            newV = V
        else:
            c = mh.cos(self._maxAngVel)
            s = mh.copysign(mh.sin(self._maxAngVel), -ang)
            newV = self._V / oldV * absV
            newV = np.array([newV[0]*c-newV[1]*s, newV[0]*s+newV[1]*c])
            #cDir = np.array([oDir[0]*c-oDir[1]*s, oDir[0]*s+oDir[1]*c])
        if abs(oldV - absV) > self._maxAcc:
            self._V = newV / absV * (oldV - mh.copysign(self._maxAcc, oldV - absV))
        else:
            self._V = newV
    def getV(self):
        return self._V
    def inN(self, RPos, RVis):
        return self.getDist(RPos) < RVis
    def isWall(self):
        return False
    
    def draw(self, sc, O):
        DispPos = self._pos*scale + O
        pygame.draw.circle(sc, self._color, (int(DispPos[0]), int(DispPos[1])), int(self._r*scale))
    def hide(self, sc, O):
        DispPos = self._pos*scale + O;
        pygame.draw.circle(sc, BGCOLOR, (int(DispPos[0]), int(DispPos[1])), int(5+self._r*scale))
    def action(self):
        self._pos = self._pos + self._V
    
class Wall(GeomObj):
    def __init__(self, A, B, width = 3, color = WALLCOLOR):
        self._A = A
        self._B = B
        self._color = color
        self._width = width
    def isLive(self):
        return True
    def getDist(self, oPos):
        v = self._A - self._B
        v = v / la.norm(v)
        oPos = oPos - self._A
        return abs(v[1] * oPos[0] - v[0] * oPos[1])
    def getYDiff(self, oPos):
        v = self._A - self._B
        return self._A[1] + (oPos[0] - self._A[0]) / v[0] * v[1] - oPos[1]
    def inN(self, RPos, RVis):
        return False
    def inRank(self, oPos, buildWidth):
        return True
    def isWall(self):
        return True
    def draw(self, sc, O):
        pygame.draw.line(sc, self._color, tuple(self._A + O), tuple(self._B + O), self._width)

class CurvedWall(Wall):
    def __init__(self, func, width = 7, color = WALLCOLOR, drawStep = 1):
        self._f = func
        self._width = width
        self._color = color
        self._drawStep = drawStep
    #norm(oDir) == 1
    def getNearestPoint(self, oPos, oDir, R = 1000, frec = 3):
        resDist = R
        eps = 1e-2
        res = oPos + RVis * oDir
        for ang in range(0, 180, frec):
            rAng = -ang / 180 * mh.pi
            c = mh.cos(rAng)
            s = mh.sin(rAng)
            cDir = np.array([oDir[0]*c-oDir[1]*s, oDir[0]*s+oDir[1]*c])
            r = self.getDirDiff(oPos, cDir, -resDist - eps, resDist + eps, eps)
            if abs(r) <= resDist - eps:
                resDist = abs(r)
                res = oPos + cDir * r
        return [resDist, res]
    def getTangentDir(self, X, h = 1e-2):
        tDir = np.array([2*h, self.getY(X+h) - self.getY(X-h)])
        return tDir / la.norm(tDir)
    def getDist(self, oPos):
        return abs(self.getYDiff(oPos)) #Вообще говоря, реальное расстояние до кривой нам не особо интересно
        #return min(la.norm(self.getNearestPoint(oPos, np.array([ 1, 0]), 100, 18)-oPos), 
        #           la.norm(self.getNearestPoint(oPos, np.array([-1, 0]), 100, 18)-oPos))
    def getDirDiff(self, oPos, oDir, l = -1000, r = 1000, eps = 1e-2):
        if oPos[1] + oDir[1] * r > self._f(oPos[0] + oDir[0] * r) and oPos[1] + oDir[1] * l > self._f(oPos[0] + oDir[0] * l):
            return r
        while r - l > eps:
            m = (l + r) / 2
            if oPos[1] + oDir[1] * m <= self._f(oPos[0] + oDir[0] * m):
                r = m
            else:
                l = m
        return r
#     def getNearestPointFast(self, oPos, oDir, R = 1000, frec = 3):
#         while r - l > eps:
#             m = (l + r) / 2
#             if oPos[1] + oDir[1] * m <= self._f(oPos[0] + oDir[0] * m):
#                 r = m
#             else:
#                 l = m
#         resDist = R
#         res = oPos + RVis * oDir
#         for ang in range(0, 180, frec):
#             rAng = -ang / 180 * mh.pi
#             c = mh.cos(rAng)
#             s = mh.sin(rAng)
#             cDir = np.array([oDir[0]*c-oDir[1]*s, oDir[0]*s+oDir[1]*c])
#             eps = 1e-2
#             r = self.getDirDiff(oPos, cDir, -R, R, eps)
#             if abs(r) <= resDist - eps:
#                 resDist = abs(r)
#                 res = oPos + cDir * r
#         return [resDist, res]
    def isDirAbove(self, oPos, oDir, l = -1000, r = 1000, eps = 1e-2):
        return self.getDirDiff(oPos, oDir, l, r, eps) < 0
    def getYDiff(self, oPos):
        return self.getY(oPos[0]) - oPos[1]
    def getY(self, X):
        return self._f(X)
    def drawSeg(self, sc, O, L, R, drawStep):
        X = L
        Y = self._f(X)
        while X < R:
            nX = X + drawStep
            nY = self._f(nX)
            pygame.draw.line(sc, self._color, (X*scale + O[0], Y*scale + O[1]), (nX*scale + O[0], nY*scale + O[1]), int(scale*self._width))
            X = nX
            Y = nY
    def draw(self, sc, O):
        self.drawSeg(sc, O, -5*WIN_WIDTH-O[0], 6*WIN_WIDTH - O[0], self._drawStep)
        
class Obstacle(Wall):
    def __init__(self, upFunc, downFunc, L, R, width = 7, color = WALLCOLOR, drawStep = 1):
        if upFunc(L) != downFunc(L) or upFunc(R) != downFunc(R):
            print('Ошибка при создании объекта. Верхняя граница не согласована с нижней')
        self._uF = upFunc
        self._dF = downFunc
        self._L = L;
        self._R = R;
        self._width = width
        self._color = color
        self._drawStep = drawStep
    def getDist(self, oPos):
        if oPos[0] < self._L:
            return la.norm(oPos - np.array([self._L, self._uF(self._L)]))
        elif oPos[0] > self._R:
            return la.norm(oPos - np.array([self._R, self._uF(self._R)]))
        else:
            return abs(self.getYDiff(oPos))
    def getXDiff(self, oPos):
        if self._L < oPos[0]:
            return self._L - oPos[0]
        elif self._R > oPos[0]:
            return self._R - oPos[0]
        else:
            return 0
    def getYDiff(self, oPos):
        return self.getY(oPos) - oPos[1]
    def getY(self, oPos):
        if oPos[0] < self._L:
            return 100500#self._uF(self._L)
        elif oPos[0] > self._R:
            return 100500#self._uF(self._R)
        uY = self._uF(oPos[0])
        dY = self._dF(oPos[0])
        return uY if abs(oPos[1] - uY) < abs(oPos[1] - dY) else dY
    def drawSeg(self, sc, O, L, R, drawStep):
        L = max(L, self._L)
        R = min(R, self._R)
        X = L
        uY = self._uF(X)
        dY = self._dF(X)
        while X < R:
            nX = X + drawStep
            nuY = self._uF(nX)
            ndY = self._dF(nX)
            pygame.draw.line(sc, self._color, (X*scale + O[0], uY*scale + O[1]), (nX*scale + O[0], nuY*scale + O[1]), int(scale*self._width))
            pygame.draw.line(sc, self._color, (X*scale + O[0], dY*scale + O[1]), (nX*scale + O[0], ndY*scale + O[1]), int(scale*self._width))
            X = nX
            uY = nuY
            dY = ndY
    def draw(self, sc, O):
        self.drawSeg(sc, O, self._L, self._R, self._drawStep)
    def isDirAbove(self, oPos, oDir): #костыль
        return self.isAbove(oPos)
    def getDirDiff(self, oPos, oDir, l = 0, r = 0, eps = 0):  #костыль
        return self.getYDiff(oPos)
    
class Scene(ActiveObj):
    def __init__(self, objs, GlobalCenter = np.array([0, 0])):
        self._objects = objs
        self._O = GlobalCenter
    def draw(self, sc):
        for obj in self._objects:
            if obj.isLive():
                obj.draw(sc, self._O)
    def action(self):
        for obj in self._objects:
            if obj.isLive():
                obj.action()  
                
class LineChart:
    def __init__(self, sc, GlobalCenter = np.array([0, 0]), FirstPnt = np.array([0, 0]), color = LINECHARTCOLOR, width = 3, Scale = np.array([1, 1])):
        self._sc = sc
        self._O = GlobalCenter
        self._prevPnt = FirstPnt
        self._color = color
        self._width = width
        self._S = Scale
    def addPnt(self, nextPnt):
        pygame.draw.line(sc, self._color, tuple(self._S*self._prevPnt+self._O), tuple(self._S*nextPnt+self._O), self._width)
        self._prevPnt = nextPnt
            
def gamma(arg, param):
    return param * arg / (1 + arg * arg) ** 0.5

def theta(arg, param):
    return param * arg

def calcL(objs, agent, RVis, buildWidth = 100000, mode = 'simple'):
    rDist = RVis
    res = None
    for obj in objs:
        cDist = abs(obj.getDirDiff(agent.getPos(), agent.getOrtDir(), -RVis-10, RVis+10)) if obj.isWall() else obj.getDist(agent.getPos())
        if obj.isLive() and cDist < RVis and obj.inRank(agent.getPos(), buildWidth):
            YDist = 2 * cDist if obj.isWall() else abs(agent.getOrtRDirDiff(obj.getPos()))
            if rDist > YDist:
                rDist = YDist
                res = obj
        #Многополосный режим
        #if obj.isLive() and obj.getDist(RPos) < RVis and obj.inRank(RPos, buildWidth):
#Old
#             YDist = obj.getYDist(RPos)
#             if rDist > YDist:
#                 rDist = YDist
#                 res = obj
    return [rDist, res]

def calcN(objs, RID):
    res = []
    for Id, obj in enumerate(objs):
        if obj.isLive() and Id != RID and obj.inN(objs[RID].getPos(), RVis):
            res.append(Id)
    return res

def forwardWallDist(objs, RPos, RVis, buildWidth, wall):
    res = 0
    for obj in objs:
        if obj.isLive() and not obj.isWall() and obj.getDist(RPos) < RVis:
            XDiff = obj.getXDiff(RPos)
            if 0.5 * buildWidth < XDiff and XDiff < 2.5 * buildWidth:
                if res == 0:
                    res = wall.getDist(obj.getPos())
                else:
                    res = min(res, wall.getDist(obj.getPos()))
    return res

def splitAboveBelow(objs, RID, rDir = np.array([0, -1])):
    resAbove = []
    resBelow = []
    for Id, obj in enumerate(objs):
        if obj.isLive() and Id != RID:
            if obj.isDirAbove(objs[RID].getPos(), rDir):
                resAbove.append(obj)
            else:
                resBelow.append(obj)
    return [resAbove, resBelow]

def getAbove(objs, RID):
    res = []
    for Id, obj in enumerate(objs):
        if obj.isLive() and Id != RID and obj.isAbove(objs[RID].getPos()):
            res.append(obj)
    return res

def getBelow(objs, RID):
    res = []
    for Id, obj in enumerate(objs):
        if obj.isLive() and Id != RID and not obj.isAbove(objs[RID].getPos()):
            res.append(obj)
    return res

def getRegLine(X, Y, cnt):
    EX = sum(X)/cnt
    EY = sum(Y)/cnt
    DX = sum((X-EX)*(X-EX))/cnt
    DY = sum((Y-EY)*(Y-EY))/cnt
    covXY = sum((X-EX)*(Y-EY))/cnt
    a = (DY-DX)/(2*covXY)
    D = mh.sqrt(1+a*a)
    c1 = a-D
    p1 = c1*EY+EX
    f1 = sum((c1*Y+X-p1)*(c1*Y+X-p1))
    c2 = a+D
    p2 = c2*EY+EX
    f2 = sum((c2*Y+X-p2)*(c2*Y+X-p2))
    if f1 < f2:
        d = np.array([c1, -1])
        return [np.array([EX, EY]), d/la.norm(d), f1]
    else:
        d = np.array([c2, -1])
        return [np.array([EX, EY]), d/la.norm(d), f2]

def checkFail(objs, rCnt):
    for obj in objs[0: rCnt]:
        RPos = obj.getPos(distortion = 0)
        if objs[rCnt].getY(RPos[0]) > RPos[1] or objs[rCnt + 1].getY(RPos[0]) < RPos[1]:
            return True
    return False

def simpleCheckConv(objs, rCnt, err = 10):
    RX = [obj.getPos()[0] for obj in objs[0: rCnt]]
    RY = [obj.getPos()[1] for obj in objs[0: rCnt]]
    RY.sort()
    RXmean = sum(RX) / rCnt
    UpY = objs[rCnt + 1].getY(RXmean)
    DownY = objs[rCnt].getY(RXmean)
    StepY = (UpY - DownY)/rCnt
    MinY = DownY + StepY * 0.5
    XDev = max([abs(X - RXmean) for X in RX])
    YDev = max([abs(MinY + j * StepY - Y) for j, Y in enumerate(RY)])
    return XDev < err and YDev < err

def checkConv(objs, rCnt, err = 10):
    RP = np.array([obj.getPos() for obj in objs[0: rCnt]])
    rl = getRegLine(RP[:, 0], RP[:, 1], rCnt)
    pnt = rl[0]
    d = rl[1]
    UpY = pnt + objs[rCnt + 1].getDirDiff(pnt, d) * d
    DownY = pnt + objs[rCnt].getDirDiff(pnt, d) * d
    StepY = (UpY - DownY)/rCnt
    MinY = DownY + StepY * 0.5
    mdist = 0
    for j in range(0, rCnt):
        cdist = min([la.norm(MinY+StepY*j-RP[k]) for k in range(0, rCnt)])
        mdist = max(mdist, cdist)
    return mdist < err

def getFormQuality(objs, rCnt):
    RP = np.array([obj.getPos() for obj in objs[0: rCnt]])
    rl = getRegLine(RP[:, 0], RP[:, 1], rCnt)
    pnt = rl[0]
    d = rl[1]
    UpY = pnt + objs[rCnt + 1].getDirDiff(pnt, d) * d
    DownY = pnt + objs[rCnt].getDirDiff(pnt, d) * d
    StepY = (UpY - DownY)/rCnt
    MinY = DownY + StepY * 0.5
    maxdist = 0
    sumdist = 0
    for j in range(0, rCnt):
        cdist = min([la.norm(MinY+StepY*j-RP[k]) for k in range(0, rCnt)])
        sumdist = sumdist+cdist
        maxdist = max(maxdist, cdist)
    return [maxdist, sumdist/rCnt]
    
def calcV(objs, N, LPlus, LMinus, V_, p1, p2, RID, buildWidth = 100000, mode = 'simple'):
    Vx = 0
    for j in N:
        #Обычный режим или режим сохранения связей
        if mode == 'simple':
            Vx = Vx + gamma(objs[j].getDirDiff(objs[RID].getPos(), objs[RID].getDir()), p1);
        #Многополосный режим
        elif mode == 'multline':
            if objs[j].inRank(objs[RID].getPos(), buildWidth):
                Vx = Vx + gamma(objs[j].getXDiff(objs[RID].getPos()), p1)
            elif abs(objs[j].getXDiff(objs[RID].getPos())) < 3 * buildWidth:
                if objs[j].getPos()[0] < objs[RID].getPos()[0]:
                    Vx = Vx + gamma(objs[j].getXDiff(objs[RID].getPos() - np.array([2 * buildWidth, 0])), p1)
                else:
                    Vx = Vx + gamma(objs[j].getXDiff(objs[RID].getPos() + np.array([2 * buildWidth, 0])), p1)
    if mode == 'connected':
        Vx = V_
    else:
        if len(N) > 0:
            Vx = Vx / len(N) + V_
        else:
            Vx = V_
    Vy = theta(LPlus, p2) - theta(LMinus, p2)
    return np.array([Vx, Vy])

#Вернёт координаты случайно точки круга радиуса R с центров в (OX, OY) (равномерное распределение)
def getRndCrcPnt(R, OX = 0, OY = 0):
    X = rdm.uniform(-R, R)
    Y = rdm.uniform(-R, R)
    #Генерируем случайную точку внутри квадрата пока не попадём в круг
    while X ** 2 + Y ** 2 > R ** 2:
        X = rdm.uniform(-R, R)
        Y = rdm.uniform(-R, R)
        
    return np.array([X + OX, Y + OY])

def getRndDir(coef = 0.5):
    a = (np.random.rand() - 0.5) * mh.pi * coef
    return np.array([mh.cos(a), mh.sin(a)])

def getRobotsDist(objs, pos):
    res = 100000
    for obj in objs:
        if obj.isLive():
         res = min(res, obj.getDist(pos))
    return res

def tupleIntMul(scalar, Tuple):
    return tuple([scalar * x for x in Tuple])

def tupleSum(tuple1, tuple2):
    l = min(len(tuple1), len(tuple2))    
    return tuple([tuple1[j] + tuple2[j] for j in range(0, l)])
      
#задаются константы
W = 350 #corridor width
initRCnt = 6
rCnt = 6 #robots count
minRCnt = 5 #minimum robots count
sensingPeriod = 5#5
killRisePeriod = 160
maxKillGroupSize = 4 #maximum of destroyed group
RVis = 90 #vision radius
RDis = 100500
maxV = 8 #maximum speed
V_ = 2 #desired speed
V_slow = 0
V_fast = 4
CC = 0.2#0.1 0.2
Mult = 0.5#2 0.5
bias = 0
O = np.array([0, 300])
scale = 0.5
checkCondition = False
condition = True

maxAngVel = 14*sensingPeriod
maxAcc = 0.08*sensingPeriod

scenario = 0
drawing = True
tracking = False
dirDrawing = False
plumFreq = 0
beadFreq = 20

wallTracking = False

simulationCount = 1000
maxrunning = 960 #960 для scale = 0.5 440 для scale = 1
buildWidth = RVis / 5;
MaxLMinusDist = (RVis ** 2 - 4 * buildWidth ** 2) ** 0.5 - 10
MaxWallDist = 7 * MaxLMinusDist / 12;
MinWallDist = 3 * MaxLMinusDist / 12;

RVelHist = np.zeros((rCnt, maxrunning, 2))

#если происходит прорисовка то симуляция не должна запускаться много раз
if drawing:
    simulationCount = 1
#проверка на допустимость констант
if W / minRCnt > RVis:
    print('Плохое соотношение W, minRCnt, RVis')
if W / rCnt > RVis:
    print('Плохое соотношение W, rCnt, RVis')
    condition = False;
if condition or not checkCondition:
    #control algorithm coefficients calculation
    BB = CC * theta(RVis, 1)
    p1 = (((1 + BB ** 2) * maxV ** 2 - (BB * V_) ** 2) ** 0.5 - V_) / (1 + BB ** 2)
    p2 = Mult * CC * p1
    print([p1, p2])
    liveCnt = rCnt
    
    if condition:
        D = max((RVis ** 2 - (W / rCnt) ** 2) ** 0.5, 50)
    else:
        D = 100;
    #случайно раскидываются роботы в полосе и инициализируются стенки
    #objs = [Robot(np.array([rdm.uniform(0, D), rdm.uniform(-W/2+50-bias, W/2-50-bias)])) for j in range(0, rCnt)]
    
    #случайно раскидываются роботы в окрестности центра коридора
    #objs = [Robot(getRndCrcPnt(D / 2, D / 2, -bias)) for j in range(0, rCnt)]
    
    #инициализируем роботов с нулевыми координатами
    objs = [Robot(np.array([0, 0]), color = tupleIntMul(1/rCnt, tupleSum(tupleIntMul(j, MINROBOTCOLOR), tupleIntMul((rCnt - j), MAXROBOTCOLOR))), Dir = getRndDir() if wallTracking else np.array([1, 0]), maxAngVel = maxAngVel, maxAcc = maxAcc) for j in range(0, rCnt)]
    
    #Прямые стенки
#     objs.append(Wall(A = np.array([-10, -W / 2 - bias]), B = np.array([WIN_WIDTH + 10, -W / 2 + bias])))
#     objs.append(Wall(A = np.array([-10, W / 2 - bias]), B = np.array([WIN_WIDTH + 10, W / 2 + bias])))
    
    #Кривые параллельные стенки
        #Самое простое
    #objs.append(CurvedWall(lambda x: 45 * math.sin(2*x/100 + 3) - W/2))
    #objs.append(CurvedWall(lambda x: 45 * math.sin(2*x/100 + 3) + W/2))
    
        #Комбинация синусов (Переиодичная)
    #objs.append(CurvedWall(lambda x: 10 * (math.sin(x/7) + math.sin(x/13)) - W/2))
    #objs.append(CurvedWall(lambda x: 10 * (math.sin(x/7) + math.sin(x/13)) + W/2))
    
        #Комбинация синусов (Непериодичная)
    Amp = 13
    Phs = 0.06
    A_ = Amp*np.random.rand(6)
    P_ = Phs*(np.random.rand(6)+1)
    S_ = 2*mh.pi*np.random.rand(6)
    #objs.append(CurvedWall(lambda x: A_[0]*mh.sin(P_[0]*x+S_[0]) + A_[1]*mh.sin(P_[1]*x+S_[1]) + A_[2]*mh.sin(P_[2]*x+S_[2]) - W/2))
    #objs.append(CurvedWall(lambda x: A_[3]*mh.sin(P_[3]*x+S_[3]) + A_[4]*mh.sin(P_[4]*x+S_[4]) + A_[5]*mh.sin(P_[5]*x+S_[5]) + W/2))
    
        #Плохой тест
    #objs.append(CurvedWall(lambda x: 20 * (math.sin(x/5) + math.sin(x/7)) - W/2))
    #objs.append(CurvedWall(lambda x: 20 * (math.sin(x/5) + math.sin(x/7)) + W/2))
    
        #Кривые и смещенные
    #objs.append(CurvedWall(lambda x: 10 * math.sin(x/11) + x * 0.2 - W))
    #objs.append(CurvedWall(lambda x: 10 * math.sin(x/11) + x * 0.2))
    
        #
    #objs.append(CurvedWall(lambda x: 7 * (mh.sin(x/13) + mh.sin(x/17)) - x * 0.2))
    #objs.append(CurvedWall(lambda x: 7 * (mh.sin(x/13) + mh.sin(x/17)) - x * 0.2 + W))
    
    #Кривые сужающиеся расширяющиеся стенки
        #Самое простое
#     objs.append(CurvedWall(lambda x:  40 * math.sin(1*math.pi * x/200) - W/2))
#     objs.append(CurvedWall(lambda x: -40 * math.sin(1*math.pi * x/200) + W/2))
    
        #
    #objs.append(CurvedWall(lambda x:  10 * (math.sin(x/5) + math.sin(x/7)) - W/2))
    #objs.append(CurvedWall(lambda x: -10 * (math.sin(x/5) + math.sin(x/7)) + W/2))
    
        #Плохой тест
    #objs.append(CurvedWall(lambda x:  20 * (math.sin(x/5) + math.sin(x/7)) - W/2))
    #objs.append(CurvedWall(lambda x: -20 * (math.sin(x/5) + math.sin(x/7)) + W/2))
    
        #Кривые и смещённые
    #Нисходящий
    #objs.append(CurvedWall(lambda x:  10 * math.sin(x/11) + x * 0.25 - W))
    #objs.append(CurvedWall(lambda x: -10 * math.sin(x/11) + x * 0.25 ))
    
    #Восходящий
#     objs.append(CurvedWall(lambda x:  10 * math.sin(x/11) - x * 0.25 ))
#     objs.append(CurvedWall(lambda x: -10 * math.sin(x/11) - x * 0.25 + W))
    
        #
    #objs.append(CurvedWall(lambda x:  10 * (math.sin(x/17) + math.sin(x/23)) + x * 0.25 - W))
    #objs.append(CurvedWall(lambda x: -10 * (math.sin(x/17) + math.sin(x/23)) + x * 0.25))
    
    #objs.append(CurvedWall(lambda x: 10 * (math.sin(x/7) + math.sin(x/13)) - W/2))
    #objs.append(CurvedWall(lambda x: 10 * (math.sin(x/7) + math.sin(x/13)) + W/2))
    
    #Streight tapering walls
#     w = 100
#     objs.append(CurvedWall(lambda x: -w/2 - x * 0.1))
#     objs.append(CurvedWall(lambda x:  w/2 + x * 0.1))
    
    #Streight walls
#     objs.append(CurvedWall(lambda x: -W/2))
#     objs.append(CurvedWall(lambda x: +W/2))
    
    #Steight offset walls
    #objs.append(CurvedWall(lambda x: -x * 0.4))
    #objs.append(CurvedWall(lambda x: -x * 0.4 + W))
    
    #Sinus walls
    objs.append(CurvedWall(lambda x: 25 * math.sin(1.5*x/100) - W/2))
    objs.append(CurvedWall(lambda x: 25 * math.sin(1.5*x/100) + W/2))
    
    #Sinus sum offset walls
#     objs.append(CurvedWall(lambda x:  11 * (math.sin(x/29) + math.sin(x/11)) + x * 0.2 - W + 100))
#     objs.append(CurvedWall(lambda x: -11 * (math.sin(x/29) + math.sin(x/11)) + x * 0.2 + 100))
    
    #_____      _____
    #     \____|
    #      ____  
    #_____|    \_____
    OX = WIN_WIDTH/2
    aa = 200
    #aa = 155
    #bb = 200
    bb = 300
    #cc = 1.5
    cc = 0.3
    #w = 215
    w = 175
#     objs.append(CurvedWall(lambda x: -cc*(abs(x-aa-OX)+abs(x+aa-OX)-abs(x-bb-OX)-abs(x+bb-OX))-w))
#     objs.append(CurvedWall(lambda x:  cc*(abs(x-aa-OX)+abs(x+aa-OX)-abs(x-bb-OX)-abs(x+bb-OX))+w))
    
    #Angle walls
    P = 120
    a = WIN_WIDTH/2
#     objs.append(CurvedWall(lambda x: P * (1 - abs(x-a) / a) - W + W/4))
#     objs.append(CurvedWall(lambda x: P * (1 - abs(x-a) / a) + W/4))
    
    #Arc walls
    R = 900
    a = WIN_WIDTH/2
    y0 = mh.sqrt(R ** 2 - a ** 2)
#     objs.append(CurvedWall(lambda x: y0 + WIN_HEIGHT/2 - W - mh.sqrt(R ** 2 - (x - a) ** 2)))
#     objs.append(CurvedWall(lambda x: y0 + WIN_HEIGHT/2 - mh.sqrt(R ** 2 - (x - a) ** 2)))
    
    #Arc plus sinus walls
    AA = WIN_WIDTH/2 + 50
    BB = WIN_HEIGHT/2
    #objs.append(CurvedWall(lambda x: -BB/AA*mh.sqrt(AA**2-((x-WIN_WIDTH/2)/K)**2) + 11*mh.sin(2**0.5*x/31) + 15*mh.sin(x/13) + W/4))
    #objs.append(CurvedWall(lambda x: -BB/AA*mh.sqrt(AA**2-((x-WIN_WIDTH/2)/K)**2) + 11*mh.sin(2**0.5*x/31) + 15*mh.sin(x/13) + 5*W/4))
    
    #objs.append(CurvedWall(lambda x: -BB/AA*mh.sqrt(AA**2-((x-WIN_WIDTH/2))**2) + 9*mh.sin(2**0.5*x/31) + 11*mh.sin(x/13) + W/4))
    #objs.append(CurvedWall(lambda x: -BB/AA*mh.sqrt(AA**2-((x-WIN_WIDTH/2))**2) - 9*mh.sin(2**0.5*x/31) - 11*mh.sin(x/13) + 5*W/4))
    
    #Chaotic walls
#     a = WIN_WIDTH/2
#     objs.append(CurvedWall(lambda x: 40 * mh.asin(mh.sin((x + 170)/200)) +17 * mh.sin(x/201) * mh.sin(x/17) - W/2))
#     objs.append(CurvedWall(lambda x: 20 * mh.asin(mh.sin(x/100)) + 13 * mh.sin(x/201) * mh.sin(x/17) + 5 * mh.asin(mh.sin(x/20)) + W/2))
    
    #Zigzag walls
#     objs.append(CurvedWall(lambda x: 30 * mh.asin(mh.sin((x + 170)/80)) - W/2))
#     objs.append(CurvedWall(lambda x: 30 * mh.asin(mh.sin((x + 170)/80)) + W/2))
    
    TT = np.arange(0, 10, 0.025)
    XX = [[] for j in range(0, rCnt)]
    LPM = [[] for j in range(0, rCnt)]
    
    #Simulation initialization
    if drawing:
        pygame.init()

        clock = pygame.time.Clock()

        sc = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))

        sc.fill(BGCOLOR)
    sm = Scene(objs, O)
    while simulationCount > 0:
        simulationCount = simulationCount - 1
        #Robots placement at the beggining of the corridor
        AverageX = 0
        SpawnR = 40
        botDist = 20
        for j in range(0, rCnt):
            position = getRndCrcPnt(SpawnR, SpawnR, (objs[rCnt].getY(D/2) + objs[rCnt+1].getY(D/2))/2)
            while getRobotsDist(objs, position) < botDist:
                position = getRndCrcPnt(SpawnR, SpawnR, (objs[rCnt].getY(D/2) + objs[rCnt+1].getY(D/2))/2)
            objs[j].teleportation(position)
        
        reserveGroup = rdm.sample(range(rCnt), rCnt-initRCnt)
        for j in reserveGroup:
            objs[j].kill()            
            if scenario == 2:
                objs[j].setState('fast')
            elif scenario == 3:
                objs[j].setState('slow')
                objs[j].teleportation(objs[j].getPos(distortion = 0)+400*np.array([1, 0]))
            objs[j].draw(sc, O)
    #         AverageX = AverageX + objs[j].getPos()[0]
    #     AverageX = AverageX / rCnt
    #     XPosCharts = [LineChart(sc, np.array([50, WIN_HEIGHT-150]), np.array([0, AverageX-objs[j].getPos()[0]]), width = 2, Scale = np.array([1, 3])) for j in range(0, rCnt)]
    #     for j in range(0, rCnt):
    #         XX[j].append(AverageX - objs[j].getPos()[0])

        #Robots direction calculation initialization
        forg = 0.991 #Forgetting coeffitient
        sCoef = 0
        V_t = [sCoef/(1-forg) * objs[j].getDir() for j in range(0, rCnt)]

        #Wall drawing
        if drawing:
            objs[rCnt].draw(sc, O)
            objs[rCnt + 1].draw(sc, O)

        #Obstacles initialization
        R = 3500
        xC = 700 
        xR = 500 #<R
        yC = mh.sqrt(R ** 2 - xR ** 2)
        yoff = 0.1
        fColl = lambda x: (10*mh.sin(7**0.5*x/39) + 11*mh.sin(3**0.5*x/37) + 7*mh.sin(2**0.5*x/29))/3;
        objs.append(Obstacle(lambda x: (x-xC)*yoff-yC+mh.sqrt(R**2-(x-xC)**2) + fColl(x), lambda x: (x-xC)*yoff+yC-mh.sqrt(R**2-(x-xC)**2) + fColl(x), xC-xR, xC+xR))

        #Obstacles drawing
        objs[rCnt + 2].draw(sc, O)

        #симуляция
        #time.sleep(5)
        running = 1
        sumquality = np.array([0, 0])
        convtime = np.array([-1.0, -1.0, -1.0, -1.0])
        connections = np.zeros(rCnt)
        hSz = 300
        hDelay = 250
        Fail = False
        if drawing:
            oldRPos_ = [R.getPos(distortion = 0) for R in objs[0: rCnt]]
            oldRDir_ = [R.getDir() for R in objs[0: rCnt]]
            oldRLiv_ = [R.isLive() for R in objs[0: rCnt]]
            oldRPos = [oldRPos_.copy() for j in range(0, hSz)]
            oldRDir = [oldRDir_.copy() for j in range(0, hSz)]
            oldRLiv = [oldRLiv_.copy() for j in range(0, hSz)]
            snapshotFrame = 80;
            snapshotFrameStep = 80;
        time.sleep(3)
        while running >= 0 and (running <= maxrunning or drawing):
            #sc.fill(BGCOLOR)
            if drawing:
                StartTime = time.time()

                for i in pygame.event.get():
                    if i.type == pygame.QUIT: running = -100500


            #sm.draw()
            #рисуются живые роботы и границы коридора при необходимости
                drawDist = 30
                #if running % 3 == 0:
                if tracking:
                    for j in range(0, rCnt):
                        Pnt1 = oldRPos[hDelay - 1][j]*scale + O
                        Pnt2 = oldRPos[hDelay - 2][j]*scale + O
                        #color = BLACK if running >= killRisePeriod+hDelay else objs[j].getColor()
                        color = objs[j].getColor()
                        pygame.draw.line(sc, color, Pnt1, Pnt2, 2)
                        
                        for k in reversed(range(1, hSz)):
                            oldRPos[k][j] = oldRPos[k - 1][j].copy()
                            oldRDir[k][j] = oldRDir[k - 1][j].copy()
                            
                        oldRPos[0][j] = objs[j].getPos(distortion = 0)
                        oldRDir[0][j] = objs[j].getDir()
                    if beadFreq > 0 and running % beadFreq == 0:
                        for j in range(0, rCnt):
                            DispPos = oldRPos[hSz - 1][j]*scale + O
                            DispDir = DispPos + scale*8*oldRDir[hSz - 1][j]
                            #color = BLACK if running >= killRisePeriod+hSz else objs[j].getColor()
                            color = objs[j].getColor()
                            pygame.draw.circle(sc, color, (int(DispPos[0]), int(DispPos[1])), 5)
                            if dirDrawing:
                                pygame.draw.line(sc, BLACK, (int(DispPos[0]), int(DispPos[1])), (int(DispDir[0]), int(DispDir[1])), 2)
                    if plumFreq > 0 and running > 0 and running % plumFreq == 0:
                        for j in range(0, rCnt):
                            DispPos = oldRPos_[j]*scale + O
                            pygame.draw.circle(sc, objs[j].getColor(), (int(DispPos[0]), int(DispPos[1])), int(10*scale))
                            dPos = oldRPos_[j] + scale*12 * oldRDir[j]
                            if dirDrawing:
                                pygame.draw.line(sc, BLACK, (oldRPos_[j][0]*scale + O[0], oldRPos_[j][1]*scale + O[1]), (dPos[0]*scale + O[0], dPos[1]*scale + O[1]), 3)
                            oldRPos_[j] = objs[j].getPos(distortion = 0)
                            oldRDir_[j] = objs[j].getDir()
                        snapshotFrame = snapshotFrame + snapshotFrameStep

                for j in range(0, rCnt):
                    if objs[j].isLive():
                        #Рисуется ближайшая к роботу точка препятствия
        #                 nPnt = objs[rCnt].getNearestPoint(objs[j].getPos(), np.array([1, 0]), RVis, 5)
        #                 nPnt[1] = nPnt[1] + O
        #                 pygame.draw.circle(sc, RED, (int(nPnt[1][0]), int(nPnt[1][1])), 7)
                        objs[j].draw(sc, O)
                        RPos = objs[j].getPos(distortion = 0)
                        #отрисовка стен
                        if abs(objs[rCnt].getYDiff(RPos)) < drawDist:
                            objs[rCnt].drawSeg(sc, O, RPos[0] - drawDist, RPos[0] + drawDist, 1)
                        if abs(objs[rCnt + 1].getYDiff(RPos)) < drawDist:
                            objs[rCnt + 1].drawSeg(sc, O, RPos[0] - drawDist, RPos[0] + drawDist, 1)
                        #отрисовка курса
                        dPos = RPos + 12 * objs[j].getDir()
                        if dirDrawing:
                            pygame.draw.line(sc, BLACK, (RPos[0]*scale + O[0], RPos[1]*scale + O[1]), (dPos[0]*scale + O[0], dPos[1]*scale + O[1]), 3)
                    #прорисовка препятствий
                    if abs(objs[rCnt + 2].getDist(RPos)) < drawDist:
                        objs[rCnt + 2].drawSeg(sc, O, RPos[0] - drawDist, RPos[0] + drawDist, 1)

                pygame.display.update()
                if running >= maxrunning:
                    if running == maxrunning:
                        print('simulation end')
                    running = running + 1
                    continue
            if running > 0 and checkFail(objs, rCnt):
                print('Fail')
                running = maxrunning + 1
                Fail = True
                continue
            quality = getFormQuality(objs, rCnt)
            sumquality = sumquality + quality
            for j, err in enumerate([5, 10, 15, 20]):
                if quality[0] < err and convtime[j] < 0:
                    convtime[j] = running * 0.02

            #Все перемещения происходят здесь
            numDisappear = 0
    #         AverageX = 0
            if running % sensingPeriod == 0:
                for j in range(0, rCnt):
                    if objs[j].isLive():
                        N = calcN(objs, j) #Visible robots set

                        newConnections = 0
                        for k in N:
                            if abs(objs[j].getXDiff(objs[k].getPos())) < RVis / 2:
                                newConnections = newConnections + 1
                        if (newConnections < connections[j]):
                            objs[j].setState('ordinary')
                        connections[j] = newConnections
                        if wallTracking:
                            #Direction calculation and sincronization
                            pntInfo1 = objs[rCnt].getNearestPoint(objs[j].getPos(), objs[j].getDir(), RVis, 5)
                            pntInfo2 = objs[rCnt + 1].getNearestPoint(objs[j].getPos(), objs[j].getDir(), RVis, 5)
                            #pntDist1 = objs[rCnt].getDirDiff(objs[j].getPos(), objs[j].getDir(), -RVis-10, RVis+10)
                            #pntDist2 = objs[rCnt + 1].getDirDiff(objs[j].getPos(), objs[j].getDir(), -RVis-10, RVis+10)
                            dirSum = np.array([0, 0])
                            for k in N:
                                dirSum = dirSum + objs[k].getDir()
                            lN = len(N);
                            if pntInfo1[0] < RVis:
                                V_t[j] = forg * V_t[j] + objs[rCnt].getTangentDir(pntInfo1[1][0])
                                nDir = (dirSum + V_t[j] / la.norm(V_t[j])) / (1 + lN)
                                objs[j].setDir(nDir/la.norm(nDir))
                            elif pntInfo2[0] < RVis:
                                V_t[j] = forg * V_t[j] + objs[rCnt + 1].getTangentDir(pntInfo2[1][0])
                                nDir = (dirSum + V_t[j] / la.norm(V_t[j])) / (1 + lN)
                                objs[j].setDir(nDir/la.norm(nDir))
                            elif lN > 0:
                                V_t[j] = 0
                                nDir = dirSum / lN
                                objs[j].setDir(nDir / la.norm(nDir))
                        splt = splitAboveBelow(objs, j, objs[j].getOrtDir())
                        LMinusInf = calcL(splt[1], objs[j], RVis, 100500)
                        LPlusInf = calcL(splt[0], objs[j], RVis, 100500)
                        LMinus = LMinusInf[0]
                        LPlus = LPlusInf[0]
                        #Если слева ничего нет то идём влево, иначе всё по старому
                        #if LMinus >= MaxLMinusDist:
                        #    newV = [V_, -0.5 * (maxV ** 2 - V_ ** 2) ** 0.5]
                        #else:

                        #Для вывода графика
                        #if len(LPM[j]) < len(TT):
                        #    LPM[j].append(LPlus - LMinus)
                        if (objs[j].getState() == 'fast'):
                            if (LPlusInf[1] == None or not LPlusInf[1].isWall() or LPlusInf[0] > RVis):
                                LPlus = RVis

                            if (LMinusInf[1] == None or not LMinusInf[1].isWall() or LMinusInf[0] > RVis):
                                LMinus = RVis
                        elif (objs[j].getState() == 'slow'):
                            LMinus = 0
                            LPlus = 0

                        if (objs[j].getState() == 'ordinary'):
                            newV = calcV(objs, N, LPlus, LMinus, V_, p1, p2, j, mode = 'simple')
                        elif (objs[j].getState() == 'fast'):
                            newV = calcV(objs, N, LPlus, LMinus, V_fast, p1, p2, j, mode = 'connected')
                        elif (objs[j].getState() == 'slow'):
                            newV = calcV(objs, N, LPlus, LMinus, V_slow, p1, p2, j, mode = 'connected')
                        #Acceleration for multiline mode
        #                 if LMinusInf[1] != None and LMinusInf[1].isWall() and forwardWallDist(objs, objs[j].getPos(), RVis, buildWidth, LMinusInf[1]) > MaxWallDist:
        #                     newV[0] = newV[0] * 2

                        #Замедление при подходе к стенке
                        #При наличии препятствий
                        #wallDist = min(wallDist, objs[rCnt + 2].getDist(objs[j].getPos()))
        #                 pntInfo1 = objs[rCnt].getNearestPoint(objs[j].getPos(), objs[j].getDir(), RVis, 10)
        #                 pntInfo2 = objs[rCnt + 1].getNearestPoint(objs[j].getPos(), objs[j].getDir(), RVis, 10)
        #                 if pntInfo1[0] < MinWallDist:
        #                     #newV[0] = newV[0] * min(max(0.15, wallDist / MinWallDist), 0.5)
        #                     nDir = pntInfo1[1] - objs[j].getPos()
        #                     nDir = nDir / la.norm(nDir)
        #                     proec = newV[0] * nDir[0] + newV[1] * nDir[1]
        #                     newV = newV - max(proec, 0) * nDir
        #                 elif pntInfo2[0] < MinWallDist:
        #                     nDir = pntInfo2[1] - objs[j].getPos()
        #                     nDir = nDir / la.norm(nDir)
        #                     proec = newV[0] * nDir[0] + newV[1] * nDir[1]
        #                     newV = newV - max(proec, 0) * nDir

                        absV = la.norm(newV)
                        if absV > maxV:
                            newV = newV / absV * maxV
                        d = objs[j].getDir()
                        objs[j].setV(np.array([d[0]*newV[0]-d[1]*newV[1],d[1]*newV[0]+d[0]*newV[1]]))
    #                 if (objs[j].getPos() + O)[0] > WIN_WIDTH + 15:
    #                     numDisappear = numDisappear + 1
                    #AverageX = AverageX + objs[j].getPos()[0]
            #AverageX = AverageX / rCnt
            #for j in range(0, rCnt):
            #    XPosCharts[j].addPnt(np.array([running, AverageX-objs[j].getPos()[0]]))
    #         for j in range(0, rCnt):
    #             if len(XX[j]) < len(TT):
    #                 XX[j].append(AverageX - objs[j].getPos()[0])

            if running >= 0 and running <= maxrunning:
                for j in range(rCnt):
                    RVelHist[j][running] = objs[j].getV();
            if running == killRisePeriod: #Единоразовое событие потери/добавления новых роботов
                if scenario == 1:
                    killGroupSize = min(liveCnt - minRCnt, maxKillGroupSize)
                    for j in range(0, killGroupSize):
                        Id = np.random.randint(rCnt)
                        while not objs[Id].isLive():
                            Id = np.random.randint(rCnt)
                        liveCnt = liveCnt - 1
                        objs[Id].kill()
                        objs[Id].hide(sc, O)
                elif scenario == 2 or scenario == 3:
                    for j in range(0, rCnt):
                        if not objs[j].isLive():
                            objs[j].rise()
                            connections[j] = 0
                    liveCnt = rCnt
    
#Random killRise events
#             if running % killRisePeriod == 0:
#                 #Kills or rises robots
#                 doKill = (np.random.randint(2) == 1 or liveCnt == rCnt) and (liveCnt > minRCnt)
#                 if doKill:
#                     #Kills robots
#                     #killGroupSize = np.random.randint(min(liveCnt - minRCnt + 1, maxKillGroupSize + 1))
#                     killGroupSize = min(liveCnt - minRCnt, maxKillGroupSize)
#                     print('Killed: ' + str(killGroupSize))
#                     for j in range(0, killGroupSize):
#                         Id = np.random.randint(rCnt)
#                         while not objs[Id].isLive():
#                             Id = np.random.randint(rCnt)
#                         liveCnt = liveCnt - 1
#                         objs[Id].kill()
#                         objs[Id].hide(sc, O)
#                 else:
#                     #Rises robots
#                     risePnt = np.array([0, 0])
#                     for j in range(0, rCnt):
#                         if objs[j].isLive():
#                             risePnt = risePnt + objs[j].getPos()
#                     risePnt = risePnt / liveCnt
#                     riseCnt = 0
#                     for j in range(0, rCnt):
#                         if not objs[j].isLive():
#                             doRise = np.random.randint(2) == 0 #вероятность воскресить 1/2
#                             if doRise:
#                                 #Воскрешение в пределах полосы
#                                 #X = rdm.uniform(risePnt[0] - D/2, risePnt[0] + D/2)
#                                 #Y = rdm.uniform(risePnt[1] - W/2 + 50, risePnt[1] + W/2 - 50)
#                                 #objs[j].teleportation(np.array([X, Y]))

#                                 #Воскрешение в окресности центра строя
#                                 position = getRndCrcPnt(D/2, risePnt[0] + 300, risePnt[1])
#                                 while getRobotsDist(objs, position) < 20:
#                                     position = getRndCrcPnt(D/2, risePnt[0] + 300, risePnt[1])
#                                 objs[j].teleportation(position)
#                                 objs[j].rise()
#                                 objs[j].setState('slow')
#                                 connections[j] = 0
#                                 riseCnt = riseCnt + 1
#                     liveCnt = liveCnt + riseCnt
#                     print('Risen:  ' + str(riseCnt)) 

            #для цикличности "телепортируем" роботов в начало коридора после того как все они заедут за границу
            #наблюдаемой зоны
            #Просто обнуляем координаты по X (ну ещё и смещение учитывается, но не особо что-то даёт в общем случае)
            #if numDisappear >= liveCnt:
            #    for j in range(0, rCnt):
            #        if objs[j].isLive():
            #           objs[j].teleportation(objs[j].getPos() - np.array([WIN_WIDTH, 2 * bias]))

            #Смотрим на расстояние до стеночки каждого робота и телепортируем так, чтобы оно сохранялось
    #         if numDisappear >= liveCnt:
    #             for j in range(0, rCnt):
    #                 if objs[j].isLive():
    #                     tpX = objs[j].getPos()[0] - WIN_WIDTH
    #                     tpY = objs[rCnt].getY(tpX) - objs[rCnt].getYDiff(objs[j].getPos())
    #                     objs[j].teleportation(np.array([tpX, tpY]))

            #стираем роботов
            if drawing:
                for j in range(0, rCnt):
                    if objs[j].isLive():
                            objs[j].hide(sc, O)

            sm.action()

            running = running + 1
            if drawing:
                SleepTime = SForFrame + StartTime - time.time()
                if SleepTime > 0:
                    time.sleep(SleepTime)
            #clock.tick(FPS)
        if drawing:
            pygame.quit()
        avgquality = sumquality/maxrunning
        if simulationCount % 50 == 0:
            results.to_csv('./experiments_log.csv', index = False, header = True)
#         details = ''
#         for j in range(0, 6):
#             details = details+' A'+str(j)+': '+str(A_[j])+' P'+str(j)+': '+str(P_[j])+' S'+str(j)+': '+str(S_[j])
        print(avgquality, convtime, 'progress: ', simulationCount)
        results = results.append({'Fail': Fail, 'avgErr': avgquality[1], 'avgMaxErr': avgquality[0], 'less5cmErrTime': convtime[0], 'less10cmErrTime': convtime[1], 'less15cmErrTime': convtime[2], 'less20cmErrTime': convtime[3]}, ignore_index = True)
        A_ = Amp*np.random.rand(6)
        P_ = Phs*(np.random.rand(6)+1)
        S_ = 2*mh.pi*np.random.rand(6)
    #инициализация графиков
    #fig1 = plt.figure()
    #fig2 = plt.figure()

    #ConvX = fig1.add_subplot(111)
    #ConvY = fig2.add_subplot(111)
    #Вывод графиков
    #Координата по X, относительно среднего арифметического абсолютных координат X роботов
    #if len(TT) == len(XX[0]):
    #    for j in range(0, rCnt):
    #        ConvX.plot(TT, np.array(XX[j]), 'k')
    
    #Разница расстояний между левым и правым ближайшими объектами
    #if len(TT) == len(LPM[0]):
    #    for j in range(0, rCnt):
    #        ConvY.plot(TT, np.array(LPM[j]), 'k')
    
    #plt.show()
```
