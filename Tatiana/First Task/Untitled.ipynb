```python
import pandas as pd
auto = pd.read_csv('auto-mpg.data', sep=' ', na_values='?')
```

```python
auto.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
head(auto)
```

```python
auto.isna().sum()
```

```python
auto[auto.horsepower.isna()]
```

```python
auto.horsepower.mean()
```

```python
auto.horsepower.fillna(auto.horsepower.mean(), inplace=True)
```

```python
auto.horsepower.isna().sum()
```

```python
auto = pd.get_dummies(auto, columns=["cylinders"] )
```

```python
auto = pd.get_dummies(auto, columns=["model year", "origin"] )
head(auto)
```

```python
from sklearn import datasets
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
y = auto.mpg
```

```python
auto.drop(columns = ["mpg", "car name"], inplace = True)
```

```python
head(auto)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(auto, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
model2 = SGDRegressor(penalty='l2', alpha=1, max_iter=10000, early_stopping=True, eta0=0.0000001)
model2.fit(X_train, y_train)
```

```python
model2.predict(X_test)
```

```python
model2.coef_
```

```python
y_pred = model2.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
y_pred
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'penalty': ['l1', 'l2'], 'alpha': [0.01, 0.1, 0.5, 1], 'max_iter': [1000, 5000, 10000]}
```

```python
model3 = SGDRegressor(early_stopping=True, eta0=0.0000001)
```

```python
gs_model = GridSearchCV(model3, parameters, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
```

```python
gs_model.fit(X_train, y_train)
```

```python
gs_model.best_params_
```

```python
gs_model.best_estimator_
```

```python
gs_model.best_score_
```
