```python
import numpy as np
import pandas as pd
from sklearn import datasets

train = pd.read_csv('drugLibTrain_raw.tsv', sep='\t', na_values='?')
test = pd.read_csv('drugLibTest_raw.tsv', sep='\t', na_values='?')
```

```python
train.drop(columns = ['Unnamed: 0', 'urlDrugName', 'condition', 'benefitsReview', 'sideEffectsReview', 'commentsReview'], inplace = True)
test.drop(columns = ['Unnamed: 0', 'urlDrugName', 'condition', 'benefitsReview', 'sideEffectsReview', 'commentsReview'], inplace = True)
```

```python
train = pd.get_dummies(train, columns = ['effectiveness', 'sideEffects'])
test = pd.get_dummies(test, columns = ['effectiveness', 'sideEffects'])
```

```python
y_train = train.rating
y_test = test.rating
train.drop(columns = ['rating'], inplace = True)
test.drop(columns = ['rating'], inplace = True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
linRegModel = LinearRegression()
linRegModel.fit(train, y_train)
y_pred = linRegModel.predict(test)
print(mean_squared_error(y_test, y_pred))
```
